# Plot VS. Scatter 

## Let us show the difference between two pyplot methods - plot() and scatter() 
## For this, we need to show some data. Let us the average height of a person from Eastern Europe with time (taken out of the head), for example. 

```python
# creating an array list of years
time = [1896, 1900, 1940, 1980]
# creating an array of height (in cm), corresponding to a given age
height = [164, 167, 173, 176]
```

## Let us use plot() and scatter() methods and see what is different

### We will see that plot() connects the points and scatter() scatters points around

```python
fig = plt.figure(figsize=[5, 5])
plt.plot(time, height)

fig = plt.figure(figsize=[5, 5])
plt.scatter(time, height)
```

---

# What about title, labels of axes? 

## We can use the methods title(), xlabel() and ylabel() and pass the string of text

## To change a font size use fonsize = N

```python
fig = plt.figure(figsize=[10, 8])
plt.scatter(time, height)

# Here goes all the plot design
plt.title("Average height VS. time", fontsize=20)
plt.xlabel("time", fontsize=30)
plt.ylabel("height", fontsize=30)
```

---

# What about the numbers (ticks) on axes?
## We use the function tick_params() with an argument of labelsize = N
## A good practice is to have ticks smaller than label size

```python
fig = plt.figure(figsize=[10, 8])
plt.scatter(time, height)

# Here goes all the plot design
plt.title("Average height VS. time", fontsize=20)
plt.xlabel("time", fontsize=30)
plt.ylabel("height", fontsize=30)
plt.tick_params(labelsize=15)
```

---

# What about the point color and size? 

## The scatter point size is controlled with a parameter "s" and color with the parameter "c" as follows

```python
fig = plt.figure(figsize=[10, 8])
# changing the scatter points color and size
plt.scatter(time, height, c="red", s=250)

# Here goes all the plot design
plt.title("Average height VS. time", fontsize=20)
plt.xlabel("time", fontsize=30)
plt.ylabel("height", fontsize=30)
plt.tick_params(labelsize=15)
```

--- 

# What about a legend? 

## To create a legend we will use a legend() method of the pyplot (plt)

## But to display something in the legend we need to have some text - it is craeted with the parameter label="text" of the scatter method as follows

```python
fig = plt.figure(figsize=[10, 8])
# changing the scatter points color and size
plt.scatter(time, height, c="red", s=250, label="Red circles")

# Here goes all the plot design
plt.title("Average height VS. time", fontsize=20)
plt.xlabel("time", fontsize=30)
plt.ylabel("height", fontsize=30)
plt.tick_params(labelsize=15)

plt.legend()
```

# To change a location of the legend, use parameter "loc"

## To change a font size of the text in the legend use fontsize=N parameter of the legend method

```python
fig = plt.figure(figsize=[10, 8])
# changing the scatter points color and size
plt.scatter(time, height, c="red", s=250, label="Red circles")

# Here goes all the plot design
plt.title("Average height VS. time", fontsize=20)
plt.xlabel("time", fontsize=30)
plt.ylabel("height", fontsize=30)
plt.tick_params(labelsize=15)

plt.legend(loc="best", fontsize=15)
```

# 2) Adjust different settings - Some interesting things 

## - Change a symbol of the scatter marker

## - Change the padding space between the title and the figure body, axes labels and the figure body.

---

## A full list of available markers can be found [on the official Matplotlib webpage](https://matplotlib.org/3.1.0/api/markers_api.html) 

## For this example, I am using a star marker as marker="*"

```python
fig = plt.figure(figsize=[10, 8])
# changing the scatter points color and size
plt.scatter(time, height, c="blue", s=1000, label="Red circles", marker="*")

# Here goes all the plot design
plt.title("Average height VS. time", fontsize=20)
plt.xlabel("time", fontsize=30)
plt.ylabel("height", fontsize=30)
plt.tick_params(labelsize=15)

plt.legend(loc="best", fontsize=15)
```

## A padding syntaxis is a bit different between title position, axes labels (xlabel, ylabel) and axes ticks


## 1) To move title up-down use additional parameter "y=N" and to move left-right use x=N
## Note: usually it is enough to move title label only up-down. Also, default value is 1.0

## 2) To move axes labels use parameter labelpad=N

## 3) Finally, to move tick values (numbers) use pad=N
## Note: pad parameter shows the pixel size betwen the tick values and the plot body

## as follows

```python
fig = plt.figure(figsize=[10, 8])
# changing the scatter points color and size
plt.scatter(time, height, c="blue", s=1000, label="Red circles", marker="*")

# Here goes all the plot design
plt.title("Average height VS. time", fontsize=20, y = 1.2)
plt.xlabel("time", fontsize=30, labelpad=50)
plt.ylabel("height", fontsize=30, labelpad=150)
plt.tick_params(labelsize=15, pad = 30)

plt.legend(loc="best", fontsize=15)
```

# 3) Making subplots

## To create subplot(s) on the figure we use the method called .subplots()
## By passing different parameters into the subplot method, one can control the number of subplots, their distribution, axes design etc.

## For more information one can visit the official page of [matplotlib on subplots](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.subplots.html?highlight=subplots#matplotlib.pyplot.subplots)

# For this section I decided to craete a new data set which might be useful to demonstrate different subplots

## Let us craete an array of numbers from 0 to 100 using a method "linspace"
## Next, we will be assigning different power laws 
- y1 = x^1
- y2 = x^2
- y3 = x^3
...
etc

```python
# we will need to import one additional standard package called numpy (for numerical python)
import numpy as np
# and as usual, importing the matplotlib library
import matplotlib.pyplot as plt
```

```python
# Creating a linearly spaced numbers from 1 to 100
# np.linspace(start, end, number of points)
x = np.linspace(1, 100, 100)

print(x)
```

```python
# Craeting the power orders
y1 = x**1
y2 = x**2
y3 = x**3
y4 = x**4
y5 = x**5
y6 = x**6
y7 = x**7
y8 = x**8
```

```python
# Create a figure
fig = plt.figure()

# Creates a subplot - no arguments/parameters in the subplot means we are craeting one subplot
ax = fig.subplots()
# using ax variable to refer to the subplot, we can plot our power law
ax.scatter(x, y1)
ax.set_title('One subplot')
```

# Creating two subplots

## side by side

## Approach 1: using subplots
### this will create (two) subplots simultaneously - useful when you have to create only few subplots (say, less than 4)

```python
# Create a figure
fig = plt.figure()

# Creates two subplots and unpacks the output array immediately
ax1, ax2 = fig.subplots(1, 2)
ax1.scatter(x, y1)
ax1.set_title('Title 1')
ax2.scatter(x, y2)
ax2.set_title('Title 2')
```

## Approach 2: using plt method .subplot 
### this will create subplots one by one, manually - useful when you have to create a number of suplots and would like to manually assign the positions of each subplot

## Syntaxis:
## ax = plt.subplot(nrows, ncols, index)

## For example, to recraete the above configuration, we need to pass nrows=1, ncols=2 and different index to two axes as following

```python
# Create a figure
fig = plt.figure()

# Creates two subplots and unpacks the output array immediately
ax1 = plt.subplot(1, 2, 1)
ax1.scatter(x, y1)
ax1.set_title('Title 1')

ax2 = plt.subplot(1, 2, 2)
ax2.scatter(x, y2)
ax2.set_title('Title 2')
```

## One plot above another 
## By analogy, we pass nrows=1, nclos=2 and different index

```python
# Create a figure
fig = plt.figure()

# Creates two subplots and unpacks the output array immediately
ax1 = plt.subplot(2, 1, 1)
ax1.scatter(x, y1)
ax1.set_title('Title 1')

ax2 = plt.subplot(2, 1, 2)
ax2.scatter(x, y2)
ax2.set_title('Title 2')
```

## To fix our knowledge and understanding of the subplots indexing let us craete a plot with 8 suplots - 4 in one row and 4 in another

```python
# Create a figure
fig = plt.figure(figsize=[10, 8])

# Creates two subplots and unpacks the output array immediately
ax1 = plt.subplot(2, 4, 1)
ax1.scatter(x, y1)
ax1.set_title('Title 1')

ax2 = plt.subplot(2, 4, 2)
ax2.scatter(x, y2)
ax2.set_title('Title 2')

ax3 = plt.subplot(2, 4, 3)
ax3.scatter(x, y3)
ax3.set_title('Title 3')

ax4 = plt.subplot(2, 4, 4)
ax4.scatter(x, y4)
ax4.set_title('Title 4')

## Second row

ax5 = plt.subplot(2, 4, 5)
ax5.scatter(x, y5)
ax5.set_title('Title 1 down')

ax6 = plt.subplot(2, 4, 6)
ax6.scatter(x, y6)
ax6.set_title('Title 2 down')

ax7 = plt.subplot(2, 4, 7)
ax7.scatter(x, y7)
ax7.set_title('Title 3 down')

ax8 = plt.subplot(2, 4, 8)
ax8.scatter(x, y8)
ax8.set_title('Title 4 down')
```

## Of course, we can use plt.subplots to assign the number of subplots simultaneously. 
## I prefer to do it one by one to get a full feeling on what is going on
---
## For example, the above code could be minimized by using for loops as following

```python

# Create a figure
fig = plt.figure(figsize=[10, 8])

ax = [0] * 8
# Now, in a loop we are going to create axes
for i in range(8):

    y = x**(i+1)
    
    ax[i] = plt.subplot(2, 4, i+1)
    ax[i].scatter(x, y)
    ax[i].set_title('Title '+str(i+1))
```

## In addition, there is a possiblity to share axes of the subplots as follows
### Share x and y axes with sharex = True and sharey = True

```python
# Create a figure
fig = plt.figure()

# Share a Y axis with each row of subplots
fig.subplots(2, 2, sharex = True, sharey=True)
```
